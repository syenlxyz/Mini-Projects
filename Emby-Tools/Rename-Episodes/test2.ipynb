```python
from pprint import pprint
import re
```

```python
file_path = 'test1.txt'
with open(file_path, 'r') as file:
    text = file.read()
    lines = text.split('\n')
len(lines)
```

```python
results = []
for line in lines:
    patterns = [
        r'\[E\d+\]',
        r'\[\d+\]',
        r'\[\d+\s?END\]',
        r'\[\d+v\d\]',
        r'\[\d+_\d+\]',
        r'\[\d+\.\d\]',
        r'\[\d+\.\dv\d\]',
        r'-\s?\d+',
        r'\.\d+',
        r'第\d+话',
        r'第\d+話'
    ]
    for pattern in patterns:
        result = re.search(pattern, line)
        if result:
            text = result.group()
            search = re.search(r'\d+(\.\d)?', text)
            results.append({pattern: [text, search.group()]})
            break
    if not result:
        print(line)
```

```python
print(len(results))
pprint(results)
```
